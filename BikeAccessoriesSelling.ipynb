3 which country have the highest profit & unitprice?

```python

pivot1=pd.pivot_table(data=df,index='Country',values=('Unit_Price','Profit'),aggfunc='sum')
print(pivot1)
```

Highest Profit: The United States leads with a profit of 11,073,644, outperforming other countries.
Highest Unit Price: Australia ranks highest with a unit price of 14,651,442.

11.Highest product selled in 2011,2012,2013,2014,2015,2016?

```python

df_filtered = df[df['Year'].isin([2011, 2012, 2013, 2014, 2015, 2016])]
df_grouped = df_filtered.groupby(['Year', 'Product'])['Order_Quantity'].sum().reset_index().sort_values(by=['Year', 'Order_Quantity'], ascending=False)
top_products_per_year = df_grouped.groupby('Year').head(1)
print(top_products_per_year)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Year',y='Order_Quantity',data=top_products_per_year,color='skyblue')
plt.title('Highest-Selling Products (2011-2016)')
plt.xticks(rotation=45,ha='right') 
plt.show()
```

From 2013 to 2016, the top-selling product was the Water Bottle - 30 oz., with peak sales of 48,102 units in 2014.
In 2011 and 2012, the highest-selling product was the Road-750 Black, 48, with 264 units sold in 2011 and 236 units in 2012.
This showcases a shift in product demand over the years, with the Water Bottle - 30 oz. dominating recent years.

12.Which  product category generates the highest overall profit?

```python

profit_by_category = df.groupby('Product_Category')['Profit'].sum()
top_profitable_categories = profit_by_category.sort_values(ascending=False).head(5)
print("Top 5 Product Categories with Highest Overall Profit:")
print(top_profitable_categories)
```

```python
plt.figure(figsize=(10, 6))
top_profitable_categories.plot(kind='bar', color='green')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.title('Top Product Categories by Profit')
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

The Bikes category generates the highest profit, with a total of 20,519,276, significantly outperforming other categories.
Accessories follows with a profit of 8,862,377, while Clothing has the lowest profit at 2,839,447.
This highlights that Bikes is the most profitable product category overall.

14,What are the top 5 best-selling products by quantity and revenue?

```python
# Group by 'Product' and sum up the 'Order_Quantity'
top_products_by_quantity = df.groupby('Product')['Order_Quantity'].sum()

# Sort in descending order to get the highest quantities and select the top 5
top_5_products_by_quantity = top_products_by_quantity.sort_values(ascending=False).head(5)


# Display the result
print("\nTop 5 Best-Selling Products by Order Quantity:")
print(top_5_products_by_quantity)
```

```python
# Group by 'Product' and sum up the 'Revenue'
top_products_by_revenue = df.groupby('Product')['Revenue'].sum()

# Sort in descending order to get the highest revenue and select the top 5
top_5_products_by_revenue = top_products_by_revenue.sort_values(ascending=False).head(5)

# Display the result
print("\nTop 5 Best-Selling Products by Revenue:")
print(top_5_products_by_revenue)
```

```python


# Pie chart for top 5 products by order quantity
plt.figure(figsize=(10, 5))

# Top products by quantity
plt.subplot(1, 2, 1)
plt.pie(top_5_products_by_quantity, labels=top_5_products_by_quantity.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Products by Order Quantity')

# Pie chart for top 5 products by revenue
plt.subplot(1, 2, 2)
plt.pie(top_5_products_by_revenue, labels=top_5_products_by_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Products by Revenue')

# Display the charts
plt.tight_layout()
plt.show()
```

This shows that while the Water Bottle - 30 oz. is the most popular in terms of quantity, products like the Road-150 Red generate the highest revenue.
